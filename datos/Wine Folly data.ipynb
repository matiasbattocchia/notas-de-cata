# Wine Folly data

```python
import re, json
```

```python
from googletrans import Translator

translator = Translator()

def translate(text):
    return translator.translate(text, src='en', dest='es').text
```

https://winefolly.com/tips/wine-aroma-wheel-100-flavors

```python
sabores = []

for line in FLAVORS.splitlines():
    if not line: continue

    en = line.lower().strip()

    if   re.match('\s{0}\w', line):
        level_0 = en
    elif re.match('\s{4}\w', line):
        level_1 = en
    elif re.match('\s{8}\w', line):
        es = translate(en).lower()

        sabor = {
            'origen': level_0,
            'descriptor': level_1,
            'nombre_es': es,
            'nombre_en': en,
        }
    
        sabores.append(sabor)
```

```python
with open('sabores.json', 'w') as f:
    json.dump(sabores, f, ensure_ascii=False, indent=2)
```
